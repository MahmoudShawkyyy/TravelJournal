```python
from PIL import Image

new_size = (512, 512)  

def resize_image(Image_path, output_size):
    try:
        img = Image.open(image_path)
        print(f"Opened image: {image_path} with size: {img.size}")
        
        img = img.resize(output_size, Image.LANCZOS)
        print(f"Resized image to: {output_size}")
        
        return img
    except FileNotFoundError:
        print(f"Error: File not found - {image_path}")
        return None
    except Exception as e:
        print(f"Error resizing image {image_path}: {e}")
        return None

for idx, row in df1.iterrows():
    image_path = row['Image_path']  
    
    resized_image = resize_image(image_path, new_size)
    
    if resized_image:
        resized_image.save(image_path)
        print(f"Resized image saved at {image_path}")
```

```python
import pandas as pd
from PIL import Image, ImageOps

target_size = (512, 512)  

def add_padding(image, target_size):
    original_size = image.size
    print(f"Original size: {original_size}")

    delta_width = target_size[0] - original_size[0]
    delta_height = target_size[1] - original_size[1]

    padding = (delta_width // 2, delta_height // 2, delta_width - (delta_width // 2), delta_height - (delta_height // 2))

    new_image = ImageOps.expand(image, padding, fill=0)
    print(f"New size with padding: {new_image.size}")
    
    return new_image

if 'Image_path' in df1.columns:
    
    for idx, row in df1.iterrows():
        Image_path = row['Image_path'] 
        
        if pd.notna(Image_path):
            try:
                img = Image.open(Image_path)

                padded_image = add_padding(img, target_size)

                padded_image.save(Image_path)
                print(f"Padded image saved at: {Image_path}")

            except Exception as e:
                print(f"Error processing image {Image_path}: {e}")
else:
    print("لا يوجد عمود 'Image Path' في الملف.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# Load your dataset
df = pd.read_csv('synthetic_travel_dataset_balanced.csv')

# Check the dataset structure
print("Dataset Columns:", df.columns)
print(df.head())

# Ensure 'Destination' column exists
if 'Destination' not in df.columns:
    print("Error: 'Destination' column is missing from the dataset.")
else:
    print(f"Unique values in 'Destination': {df['Destination'].unique()}")

    # Separate features and target
    y = df['Destination']
    X = df.drop('Destination', axis=1)

    # One-hot encode categorical data
    X_encoded = pd.get_dummies(X, drop_first=True)

    # Check for rare classes
    class_counts = y.value_counts()
    print("\nClass Counts Before Handling Rare Classes:")
    print(class_counts)

    # Remove classes with very few samples (fewer than 6 samples)
    rare_classes = class_counts[class_counts < 6].index
    if len(rare_classes) > 0:
        print(f"\nRemoving rare classes: {list(rare_classes)}")
        X_encoded = X_encoded[~y.isin(rare_classes)]
        y = y[~y.isin(rare_classes)]

    # Check the updated class distribution
    print("\nClass Counts After Removing Rare Classes:")
    print(y.value_counts())

    # Apply SMOTE for class balancing
    try:
        smote = SMOTE(sampling_strategy='auto', random_state=42, k_neighbors=2)
        X_balanced, y_balanced = smote.fit_resample(X_encoded, y)

        # Check balanced distribution
        print("\nClass Counts After SMOTE Balancing:")
        print(pd.Series(y_balanced).value_counts())

        # Train-test split
        X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.3, random_state=42)

        # Scale numerical features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # Define and train the model
        mlp = MLPClassifier(random_state=42, early_stopping=True, validation_fraction=0.1, n_iter_no_change=10)

        param_grid = {
            'hidden_layer_sizes': [(50,), (100,)],
            'activation': ['relu', 'tanh'],
            'solver': ['adam', 'sgd']
        }

        grid_search = GridSearchCV(mlp, param_grid, cv=5, n_jobs=-1, verbose=2)
        grid_search.fit(X_train_scaled, y_train)

        # Evaluate the model
        best_mlp_model = grid_search.best_estimator_
        print(f"\nBest Parameters: {grid_search.best_params_}")

        y_pred = best_mlp_model.predict(X_test_scaled)
        print(f"\nTest Accuracy: {accuracy_score(y_test, y_pred) * 1:.2f}%")
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred))

        # Plot the confusion matrix
        conf_matrix = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(10, 7))
        sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", 
                    xticklabels=best_mlp_model.classes_, 
                    yticklabels=best_mlp_model.classes_)
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.title('Confusion Matrix')
        plt.show()

        # Save the trained model
        joblib.dump(best_mlp_model, 'best_mlp_model.pkl')
        print("\nModel saved as 'best_mlp_model.pkl'")

    except ValueError as e:
        print(f"\nError during SMOTE or training: {e}")
```

```python
import pandas as pd
import numpy as np
import joblib

# Load the trained model and scaler
model = joblib.load("C:/Users/HP/Pictures/SRSProject/project/best_mlp_model_with_scaler.pkl")  # Load the trained model


# Define the feature input prompt
def get_user_input():
    print("\nPlease provide the following input features:")

    # Take inputs from user
    type_of_city = input("Type of City (Historical/Urban/Mountain/Coastal): ").strip()
    accommodation_cost = float(input("Accommodation Cost (numeric): "))
    transportation_cost = float(input("Transportation Cost (numeric): "))
    type_of_tourism = input("Type of Tourism (Cultural/Adventure/Beach/Medical/Nature): ").strip()
    continent = input("Continent (Asia/Europe/Africa/North America/South America): ").strip()
    weather = input("Weather (Tropical/Temperate/Arid/Cold): ").strip()

    # Combine inputs into a dictionary
    features = {
        'Type of City': type_of_city,
        'Accommodation cost': accommodation_cost,
        'Transportation cost': transportation_cost,
        'Type of Tourism': type_of_tourism,
        'Continent': continent,
        'Weather': weather
    }

    return features

# Preprocess the input features
def preprocess_input(features, scaler, feature_columns):
    # One-hot encoding for categorical values
    df_input = pd.DataFrame([features])  # Convert input to DataFrame
    df_input_encoded = pd.get_dummies(df_input)

    # Reindex to match training features and fill missing columns with 0
    df_input_encoded = df_input_encoded.reindex(columns=feature_columns, fill_value=0)

    # Standardize numerical columns
    df_scaled = scaler.transform(df_input_encoded)

    return df_scaled

# Main prediction function
def main():
    # Load the training feature columns
    print("Loading model and feature columns...")
    model_data = joblib.load('best_mlp_model_with_scaler.pkl')
    model = model_data['model']
    scaler = model_data['scaler']
    feature_columns = model_data['features']

    print("Model loaded successfully!\n")

    # Get user input
    features = get_user_input()

    # Preprocess input
    try:
        input_processed = preprocess_input(features, scaler, feature_columns)

        # Make prediction
        prediction = model.predict(input_processed)
        print(f"\nPredicted Destination: {prediction[0]}")
    except Exception as e:
        print(f"Error during prediction: {e}")

if __name__ == '__main__':
    main()
```
